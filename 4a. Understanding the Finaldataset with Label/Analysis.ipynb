```python

fig, axs = plt.subplots(3, 2, figsize=(20,20))

# Flatten the array of axes so that it can be easily indexed
axs = axs.ravel()

# Set titles for each subplot
titles = ['sentiment_label_headline_negative', 'sentiment_label_headline_neutral',
       'sentiment_label_headline_positive', 'sentiment_label_content_negative',
       'sentiment_label_content_neutral', 'sentiment_label_content_positive',]

# Iterate over each column and create a histogram in the corresponding subplot
for i,col in enumerate(dff.columns):
    if i<6:
        axs[i].plot(dff[col])
        axs[i].set_title(titles[i])

# Show the figure
plt.show()
```

```python
df = pd.read_csv("/content/drive/MyDrive/CBSA22(Elaine)/2020to2022data.csv")
df
```

```python
from tqdm import tqdm
import pandas as pd
import time

df["tran_headline"] = ""
df["tran_content"] = ""
count = 0
i = 0
exceptions = []

#config
start= 65001 
end= 85001
checkpt=2000

for num in tqdm(range(start+1, end+2)):
    headline = df["headline"][num]
    content = df["content"][num]
    count += 1
    i += 1
    try:
        df["tran_headline"][num] = translator.translate(headline, dest="en").text
        df["tran_content"][num] = translator.translate(content, dest="en").text
    except Exception as e:
        df["tran_headline"][num] = ""
        df["tran_content"][num] = ""
        exceptions.append((num, e))
        print(f"Error occurred at index {num}: {e}")

    if count == 10:
        time.sleep(10)
        count = 0    
    
    if i % checkpt == 0 or num == end:
        df_senti_index = df.iloc[num-i:num]
        df_senti_index.to_csv(f"df_senti_index{num-1}.csv")
        i = 0

# Save the exceptions to a CSV file
exceptions_df = pd.DataFrame(exceptions, columns=["index", "exception"])
exceptions_df.to_csv("exceptions.csv", index=False)
```
